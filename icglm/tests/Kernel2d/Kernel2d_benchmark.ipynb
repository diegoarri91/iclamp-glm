# convolve_continuous <a name="convolve_continuous"></a>

Kernel method. Given a 1d-array t and an nd-array I with I.shape=(len(t),...) returns conv, the convolution of the kernel with axis 0 of I for all other axis values so that conv.shape = I.shape

```python
from scipy.signal import fftconvolve

from icglm.utils.time import get_dt

def convolve_continuous2(self, t, I):

    dt = get_dt(t)
    arg0x = int(self.support_x[0] / dt)
    argfx = int(np.ceil(self.support_x[1] / dt))
    arg0y = int(self.support_y[0] / dt)
    argfy = int(np.ceil(self.support_y[1] / dt))

    tx = np.arange(arg0x, argfx + 1, 1) * dt
    ty = np.arange(arg0y, argfy + 1, 1) * dt
    kernel_values = self.interpolate(tx, ty)

    kernel_values = kernel_values.reshape(kernel_values.shape + tuple([1] * (I.ndim-1)))

    conv_rows = fftconvolve(kernel_values, I[None, ...], mode='full', axes=1)[:, :len(t)]  # convolucion sobre columnas para cada fila de C
    convolution = fftconvolve(I.reshape((I.shape[0], 1) + I.shape[1:]), conv_rows, mode='full', axes=0)
    convolution = convolution[:len(t), ...]
    convolution = np.moveaxis(np.diagonal(convolution), -1, 0) * dt**2

    return convolution

def convolve_continuous3(self, t, I):

    dt = get_dt(t)
    arg0x = int(self.support_x[0] / dt)
    argfx = int(np.ceil(self.support_x[1] / dt))
    arg0y = int(self.support_y[0] / dt)
    argfy = int(np.ceil(self.support_y[1] / dt))

    tx = np.arange(arg0x, argfx + 1, 1) * dt
    ty = np.arange(arg0y, argfy + 1, 1) * dt
    # tx, ty = np.meshgrid(tx, ty)
    kernel_values = self.interpolate(tx, ty)

    kernel_values = kernel_values.reshape(kernel_values.shape + tuple([1] * (I.ndim-1)))

    conv_rows = fftconvolve(kernel_values, I[None, ...], mode='full', axes=1)[:, :len(t)]  # convolucion sobre columnas para cada fila de C

    fshape = [next_fast_len(d) for d in [len(t)]]
    sp1 = np.fft.rfftn(I.reshape((I.shape[0], 1) + I.shape[1:]), fshape, axes=0)
    sp2 = np.fft.rfftn(conv_rows, fshape, axes=0)
    convolution = np.fft.irfftn(sp1 * sp2, fshape, axes=0)

    # convolution = fftconvolve(I.reshape((I.shape[0], 1) + I.shape[1:]), conv_rows, mode='full', axes=0)
    convolution = convolution[:len(t), ...] * dt**2

    return convolution

tbins = np.arange(0, 180, 10)
tbins2d = np.meshgrid(tbins, tbins)
coefs = np.exp(-(tbins2d[0][:-1, :-1]**2 + tbins2d[1][:-1, :-1]**2)**.5/20)

KernelRect2d.convolve_continuous2 = convolve_continuous2
KernelRect2d.convolve_continuous3 = convolve_continuous3

ker = KernelRect2d(tbins_x=tbins, tbins_y=tbins, coefs=coefs)

t = np.arange(0, 1000, 1)
np.random.seed(0)
stim = np.random.randn(len(t))
conv1 = ker.convolve_continuous(t, stim)
conv2 = ker.convolve_continuous2(t, stim)
# conv3 = ker.convolve_continuous3(t, stim)

# print(np.all(np.abs(conv_true - conv) < 1e-6))

# %timeit -r3 -n1 ker.convolve_continuous(t, stim)
# %timeit -r3 -n1 ker.convolve_continuous2(t, stim)

# fig,ax = plt.subplots(figsize=(16, 5))
# ax.plot(t, conv_true)
# ax.plot(t, conv)
```

# convolve_basis_continuous <a name="convolve_basis_continuous"></a>

```python
from icglm.utils.time import get_dt, searchsorted

def convolve_basis_continuous2(self, t, I):

    # Given a 1d-array t and an nd-array I with I.shape=(len(t),...) returns X,
    # the convolution matrix of each rectangular function of the base with axis 0 of I for all other axis values
    # so that X.shape = (I.shape, nbasis)
    # Discrete convolution can be achieved by using an I with 1/dt on the correct timing values

    dt = get_dt(t)
    n_independent_coefs = (len(self.tbins_x) - 1) * len(self.tbins_x) // 2

    arg_binsx = searchsorted(t, self.tbins_x)
    arg_binsy = searchsorted(t, self.tbins_y)

    basis_shape = tuple([len(t), len(t)] + [1 for ii in range(I.ndim - 1)] + [n_independent_coefs])
    basis = np.zeros(basis_shape)

    # for k, (arg0x, argfx, arg0y, argfy) in enumerate(zip(arg_binsx[:-1], arg_binsx[1:], arg_binsy[:-1],
    #                                                      arg_binsy[1:])):
    for k, (i, j) in enumerate(zip(*np.triu_indices(len(self.tbins_x) - 1))):
        arg0x, argfx = arg_binsx[i], arg_binsx[i + 1]
        arg0y, argfy = arg_binsy[j], arg_binsy[j + 1]
        if i == j:
            basis[arg0x:argfx, arg0y:argfy, ..., k] = 1.
        else:
            basis[arg0x:argfx, arg0y:argfy, ..., k] = 2.

    conv_rows = fftconvolve(basis, I[None, ..., None], mode='full', axes=1)[:, :len(t)]
    del basis
    X = fftconvolve(I.reshape((I.shape[0], 1) + I.shape[1:] + (1,)), conv_rows, mode='full', axes=0)[:len(t), :]
    X = np.moveaxis(np.diagonal(X), -1, 0) * dt**2

    return X

def convolve_basis_continuous3(self, t, I):

    # Given a 1d-array t and an nd-array I with I.shape=(len(t),...) returns X,
    # the convolution matrix of each rectangular function of the base with axis 0 of I for all other axis values
    # so that X.shape = (I.shape, nbasis)
    # Discrete convolution can be achieved by using an I with 1/dt on the correct timing values

    # dt = get_dt(t)
    n_independent_coefs = (len(self.tbins_x) - 1) * len(self.tbins_x) // 2
    shape = (len(self.tbins_x) - 1, len(self.tbins_x) - 1)
    X = np.zeros(I.shape + (n_independent_coefs,))

    for k, (i, j) in enumerate(zip(*np.triu_indices(len(self.tbins_x) - 1))):
        _coefs = np.zeros(shape)
        _coefs[i, j] = 1
        X[..., k] = KernelRect2d(tbins_x=self.tbins_x, tbins_y=self.tbins_y, coefs=_coefs).convolve_continuous(t, I)
        if i != j:
            X[..., k] = 2 * X[..., k]

    return X

KernelRect2d.convolve_basis_continuous2 = convolve_basis_continuous2
KernelRect2d.convolve_basis_continuous3 = convolve_basis_continuous3

tbins = np.arange(0, 40, 10)
ker = KernelRect2d(tbins_x=tbins, tbins_y=tbins)

t = np.arange(0, 200, .1)
# signal = np.zeros(len(t))
# signal[500:540] = 1
np.random.seed(0)
signal = np.random.randn(len(t))

X = ker.convolve_basis_continuous(t, signal)
%timeit -r3 -n1 ker.convolve_basis_continuous(t, signal)
%timeit -r3 -n1 ker.convolve_basis_continuous2(t, signal)
%timeit -r3 -n1 ker.convolve_basis_continuous3(t, signal)

fig, (ax1, ax2) = plt.subplots(figsize=(16, 5.5), nrows=2)
ax1.plot(t, X[:, 0]);
ax1.plot(t, X[:, 3]);
ax1.plot(t, X[:, 5]);
# from icglm.masks import shift_mask
# ax1.plot(t, shift_mask(X[:, 0], 100, fill_value=0));
ax2.plot(t, X[:, 1]);
# ax2.plot(t, X[:, 2]);
ax2.plot(t, X[:, 4]);
```

# deconvolve_continuous <a name="deconvolve_continuous"></a>

```python
tbins = np.arange(0, 40, 10)
tbins2d = np.meshgrid(tbins, tbins)
coefs = np.exp(-(tbins2d[0][:-1, :-1]**2 + tbins2d[1][:-1, :-1]**2)**.5/20)
ker_true = KernelRect2d(tbins_x=tbins, tbins_y=tbins, coefs=coefs)

t = np.arange(0, 1000, 1)
signal = np.random.randn(len(t))
conv_true = ker_true.convolve_continuous(t, signal)

ker = KernelRect2d(tbins_x=tbins, tbins_y=tbins)
ker.deconvolve_continuous(t, signal, conv_true)
conv = ker.convolve_continuous(t, signal)

print(np.allclose(conv_true, conv, rtol=1e-2))
print(np.allclose(ker_true.coefs, ker.coefs, rtol=1e-3))

fig, (ax1, ax2) = plt.subplots(figsize=(16, 5), ncols=2)

cbar = ax1.pcolor(tbins, tbins, ker_true.coefs)
vmin, vmax = cbar.get_clim()
ax2.pcolor(tbins, tbins, ker.coefs, vmin=vmin, vmax=vmax)

fig, ax = plt.subplots(figsize=(16, 5))
ax.plot(t, conv_true)
ax.plot(t, conv)
```

```python
def hold_noise(t, tau, mu, sigma):
    from fun_signals import get_dt
    
    npoints = len(t)
    dt = get_dt(t)

    I = np.zeros( (npoints) ) * np.nan
    I[0] = mu
    
    ii = 0
    while ii < npoints-1:
        
        I[ii+1] = I[ii] + (mu - I[ii])/tau * dt + ( 2.*sigma**2./tau )**.5  * np.random.randn() * np.sqrt(dt)
        
        ii += 1
        
    return I

tbins = np.append(0, np.logspace(0, 2, 5) )
tbins = np.arange(0., 200., 25.)
coefs = np.exp(-tbins[:-1]/10.)
ker = KernelRect(tbins, coefs)
kerf = KernelRect(tbins)

t = np.arange(0, 20000, .1)
I = hold_noise(t, 3., 0., 10.)

v = ker.convolve(t, I)

kerf.deconvolve(t, I, v)
vf = kerf.convolve(t, I)

fig,(ax1, ax2) = plt.subplots(figsize = (8, 5), ncols = 2)

ker.plot(np.arange(0., tbins[-1], .1), ax = ax1)
kerf.plot(np.arange(0., tbins[-1], .1), ax = ax1)
ax2.plot(t, v)
ax2.plot(t, vf)
```
