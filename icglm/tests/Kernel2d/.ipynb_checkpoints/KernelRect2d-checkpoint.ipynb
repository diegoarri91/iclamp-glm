# convolve_continuous <a name="convolve_continuous"></a>

Kernel method. Given a 1d-array t and an nd-array I with I.shape=(len(t),...) returns conv, the convolution of the kernel with axis 0 of I for all other axis values so that conv.shape = I.shape

```python
def convolve_continuous_as_dot(self, t, shiftedI, sub=True):
    from icglm.utils.time import get_dt

    dt = get_dt(t)
    arg0x = int(self.support_x[0] / dt)
    argfx = int(np.ceil(self.support_x[1] / dt))
    arg0y = int(self.support_y[0] / dt)
    argfy = int(np.ceil(self.support_y[1] / dt))

    tx = np.arange(arg0x, argfx, 1) * dt
    ty = np.arange(arg0y, argfy, 1) * dt
    kernel_values = self.interpolate(tx, ty)
    if sub:
        shiftedI = shiftedI - shiftedI[..., :, 0]
        
    convolution = shiftedI @ kernel_values
    convolution = shiftedI * convolution
    convolution = np.sum(convolution, -1).T * dt**2
    
    return convolution

def convolve_continuous(self, t, I):

    from icglm.utils.time import get_dt
    from scipy.signal import fftconvolve

    arg0x = int(self.support_x[0] / dt)
    argfx = int(np.ceil(self.support_x[1] / dt))
    arg0y = int(self.support_y[0] / dt)
    argfy = int(np.ceil(self.support_y[1] / dt))

    tx = np.arange(arg0x, argfx, 1) * dt
    ty = np.arange(arg0y, argfy, 1) * dt
    kernel_values = self.interpolate(tx, ty)
    kernel_values = kernel_values.reshape(kernel_values.shape + tuple([1] * (I.ndim-1)))

    conv_rows = fftconvolve(kernel_values, I[None, ...], mode='full', axes=1)[:, :len(t)]  # convolucion sobre columnas para cada fila de C
    
    n_rows = conv_rows.shape[0]
    convolution = np.zeros(I.shape)
    for u in range(len(t)):
        if u < n_rows:
            convolution[u] = np.sum(I[u::-1] * conv_rows[:u + 1, u, ...], 0)
        else:
            convolution[u] = np.sum(I[u:u - n_rows:-1] * conv_rows[:n_rows, u, ...], 0)
    convolution = np.array(convolution) * dt ** 2
    
    return convolution
    
KernelRect2d.convolve_continuous = convolve_continuous
KernelRect2d.convolve_continuous_as_dot = convolve_continuous_as_dot

tbins = np.arange(0, 360, 1)
tbins2d = np.meshgrid(tbins, tbins)
coefs = np.exp(-(tbins2d[0][:-1, :-1]**2 + tbins2d[1][:-1, :-1]**2)**.5/20)
# coefs[np.diag_indices(coefs.shape[0])] = 0
ker = KernelRect2d(tbins_x=tbins, tbins_y=tbins, coefs=coefs)

dt = 1
t = np.arange(0, 1000, dt)

np.random.seed(0)
# stim = np.random.randn(len(t), 1)
from icglm.processes import OUProcess
stim = OUProcess(tau=3).sample(t)
# conv_true = ker.convolve_continuous2(t, stim)
# conv = ker.convolve_continuous(t, stim)

argf = int(np.ceil(ker.support_y[1] / dt))
design_matrix = np.zeros(stim.shape[1:] + (len(t), argf))
for arg in range(argf):
    design_matrix[..., arg:, arg] = stim[:len(t) - arg].T
conv_dot = ker.convolve_continuous_as_dot(t, design_matrix)
conv = ker.convolve_continuous(t, stim)

%timeit -n1 -r3 ker.convolve_continuous(t, stim)
%timeit -n1 -r3 ker.convolve_continuous_as_dot(t, design_matrix)

# print(np.all(np.abs(conv_true - conv) < 1e-6))
# print(np.all(np.abs(conv_true - conv_dot) < 1e-6))
# print(np.all(np.abs(conv_true - conv3) < 1e-6))
print(np.all(np.abs(conv_dot - conv) < 1e-6))

fig,ax = plt.subplots(figsize=(16, 5))
ax.plot(t, conv)
ax.plot(t, conv_dot)
# ax.plot(t, stim)
# ax.plot(t[:-1], stim[:-1] * stim[1:])
# ax.plot(t[:-1], stim[:-1] * (stim[1:] - stim[:-1]))
# ax.plot(t, conv_dot)
```

```python
def convolve_continuous1d(self, t, I, axis=0):
    from icglm.utils.time import get_dt
    from scipy.signal import fftconvolve

    dt = get_dt(t)
    arg0x = int(self.support_x[0] / dt)
    argfx = int(np.ceil(self.support_x[1] / dt))
    arg0y = int(self.support_y[0] / dt)
    argfy = int(np.ceil(self.support_y[1] / dt))

    tx = np.arange(arg0x, argfx + 1, 1) * dt
    ty = np.arange(arg0y, argfy + 1, 1) * dt
    kernel_values = self.interpolate(tx, ty)

    kernel_values = kernel_values.reshape(kernel_values.shape + tuple([1] * (I.ndim-1)))
    kernel_values = np.sum(kernel_values, axis=axis)
    n = len(kernel_values)
    
    convolution = np.zeros(len(t))
    
    for u in range(len(t)):
        if u < n:
            convolution[u] = kernel_values[:u + 1] @ I[u::-1] # hasta t
        elif u < len(t):
            convolution[u] = I[u:u - n:-1] @ kernel_values[:n]
    
    return convolution
    
KernelRect2d.convolve_continuous1d = convolve_continuous1d

tbins = np.arange(0, 180, 10)
tbins2d = np.meshgrid(tbins, tbins)
coefs = np.exp(-(tbins2d[0][:-1, :-1]**2 + tbins2d[1][:-1, :-1]**2)**.5/20)
ker = KernelRect2d(tbins_x=tbins, tbins_y=tbins, coefs=coefs)

t = np.arange(0, 1000, 1)

np.random.seed(0)
stim = np.random.randn(len(t))
conv_true = ker.convolve_continuous1d(t, stim)
# conv = ker.convolve_continuous(t, stim)

# print(np.all(np.abs(conv_true - conv) < 1e-6))

fig,ax = plt.subplots(figsize=(16, 5))
ax.plot(t, conv_true)
# ax.plot(t, conv)
```
